** BIENVENIDX AL JUEGO DE LAS CAPITALES **

- Cada jugador cuando empieza el juego tiene 3 vidas
- Hay 4 ranking del juego (0, 1, 2 y 3) según tus vidas, (darle un nombre ludico a cada nivel alcanzado por el jugador: Indiana Jones o Cristobal Colón, Jules Verne, etc.)
- Cada fallo resta una vida, al terminar tus vidas, sales del juego y el programa te dice el ranking alcanzado.
- A cada respuesta el programa te devuelve "Correcto, te quedan 3 vidas" o "Error, la capital correcta es XXX, te quedan 2 vidas"
- Cada tres aciertos sube la dificultad
- Son 9 preguntas en total
- Tienes 3 opciones y hay que elegir la correcta.
- Si falla una respuesta, pasas a la pregunta siguiente
- Si pierdes todas tus vidas, salta un mensaje que dice "Has perdido todas tus vidas, ¿quieres volver a empezar?"

```python
import random
    
def juego():     
 
    nombre_jugador = input(f"¿Cual es tu nombre?")
    print(f"¡Bienvenido/a, {nombre_jugador}, al emocionante juego de las capitales!.\nTienes 3 vidas para responder correctamente a 9 preguntas.\nCada tres respuestas correctas, la dificultad aumentará.\n¡Buena suerte, que comience el juego!")
   
    global preguntas_hechas, level, vidas, puntuacion  

    vidas = 3
    preguntas_totales = 9
    preguntas_por_nivel = 3
    level = 1
    puntuacion = 0

    preguntas_hechas = 0

    niveles = ["facil", "medio", "dificil"]

    while preguntas_hechas < preguntas_totales and vidas > 0:

        
        nivel_actual = niveles[level - 1]
            
        pais = random.choice(list(capitales_con_paises[nivel_actual].keys()))
        capital_correcta = capitales_con_paises[nivel_actual][pais]['capital']

        respuesta = input(f"¿Cual es la capital de {pais}?").strip()
                
                  
        if respuesta.lower() == capital_correcta.lower():
            puntuacion += 1
            print(f"¡Correcto!")
            
    
        else:
            vidas -= 1
            print(f"Incorrecto, la capital de {pais} es {capital_correcta}.")
            
        
        preguntas_hechas += 1
            
        if puntuacion % 3 == 0 and level < len(niveles):
            level += 1
           
        if puntuacion == 2:
            print(f'Eres como Don Quijote, {nombre_jugador}')
        
        elif puntuacion == 4:
            print(f'Estás llegando a la frontera, {nombre_jugador}')

        elif puntuacion == 6:
            print(f'Conquistando como Napoleón, {nombre_jugador}')

        elif puntuacion == 8:
            print(f'Enhorabuena {nombre_jugador}, o mejor dicho, ¡Willy Fog!')

        elif puntuacion == 9:
            print(f'{nombre_jugador}, ¡Hasta el infinito y más allá!')    

        
    if vidas == 0:

        print(f"\n¡Has perdido todas tus vidas, {nombre_jugador}!")
    
            
      

    else:

        print(f'¡Felicidades, {nombre_jugador}! Has terminado el juego con {vidas} vidas y {puntuacion} preguntas acertadas.')



                
     
```
